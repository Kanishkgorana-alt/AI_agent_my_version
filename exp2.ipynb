```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Churn_Modelling.csv")
```

```python
df.head()
```

```python
scale = StandardScaler()
cols = [3,6,8,9,12]
df.iloc[:, cols] = scale.fit_transform(df.iloc[:, cols])
```

```python
i, j = 11, 13  # set your indices 8,9,10,11
# Num_of_products ~ E
# Has_credit_card,active member --count
x_col = df.columns[i]
y_col = df.columns[j]

plt.figure()
scatter = plt.scatter(df[x_col], df[y_col], c=df["Exited"], alpha=0.6)
plt.xlabel(x_col)
plt.ylabel(y_col)
plt.title(f"{x_col} vs {y_col} (colored by Exited)")
plt.colorbar(scatter, label="Exited")
plt.show()
```

```python


i = 11
col = df.columns[i]
plt.figure()
sns.histplot(data=df, x=col, hue="Exited", stat="density", bins=30, common_norm=False)
plt.show()
plt.figure()
sns.kdeplot(data=df, x=col, hue="Exited", common_norm=False)
plt.title(f"Density of {col} by Exited")
plt.show()
```
